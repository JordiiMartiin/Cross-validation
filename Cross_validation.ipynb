```python
# cross-Validation simple
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Cargar datos
df = pd.read_csv('dataset.csv')
df = df.dropna(subset=['target_variable'])  # eliminar filas sin target
X = df.drop('target_variable', axis=1)
y = df['target_variable']

# Crear modelo
model = RandomForestClassifier(
    random_state=42,          # control de aleatoriedad
    class_weight='balanced',  # ajuste para clases desbalanceadas
    n_estimators=337,         # número de árboles
    max_depth=35,             # profundidad máxima
    min_samples_split=9,      # tamaño mínimo para dividir un nodo
    min_samples_leaf=5,       # muestras mínimas por hoja
    max_features=None,
    n_jobs=-1                 # uso de todos los núcleos
)

# Cross-validation con 5 folds
scores = cross_val_score(    # ejecuta validación cruzada
    model, X,                #datos de entrada
    y,                       #datos de salida
    cv=5,
    scoring='accuracy'
)

# Resultados
print("Accuracy por fold:", scores)
print("Accuracy media: %.4f" % scores.mean())
print("Desviación estándar: %.4f" % scores.std())
```
